```python
import pandas as pd
dic={'Email':['shubham@gmail.com','somesh@gmail.com','yuvraaj@gmail.com']}
df6=pd.DataFrame(dic)
df6
```

```python
uname=[]
for i in df6['Email']:
    k=0
    for j in i:
        k=k+1
        if(j=='@'):
            break
    uname.append(i[0:k-1])        
```

```python
uname
```

```python
df6['Username']=uname
df6
```
